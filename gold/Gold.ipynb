```python
import pandas as pd

# Cargar el dataset desde el archivo CSV
file_path = "/content/drive/MyDrive/btc_prediction_project/data/Daily.csv"
df_gold = pd.read_csv(file_path)

# Convertir la columna "Date" a tipo datetime para facilitar análisis temporal
df_gold["Date"] = pd.to_datetime(df_gold["Date"])

# Ordenar el DataFrame por fecha ascendente para mantener la secuencia temporal correcta
df_gold = df_gold.sort_values("Date")

# Mostrar las primeras filas para verificar carga correcta
print("Primeras filas del dataset:")
print(df_gold.head())

# Obtener la fecha mínima y máxima para conocer el rango temporal de los datos
start_date = df_gold["Date"].min().strftime("%Y-%m-%d")
end_date = df_gold["Date"].max().strftime("%Y-%m-%d")
print(f"\nFecha de inicio: {start_date}")
print(f"Fecha final:     {end_date}")

# Extraer solo la columna de fecha y el precio del oro en USD para análisis enfocado
df_gold_usd = df_gold[["Date", "USD"]].copy()

# Renombrar la columna "USD" a "Close_Gold_USD" para mayor claridad
df_gold_usd = df_gold_usd.rename(columns={"USD": "Close_Gold_USD"})

# Establecer la columna "Date" como índice para facilitar operaciones de series temporales
df_gold_usd.set_index("Date", inplace=True)

# Mostrar el DataFrame listo para análisis con índice temporal y precios en USD
print("\nDataset listo con índice temporal y precios en USD:")
print(df_gold_usd.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Si "Date" está en columnas, pasarlo a índice
if "Date" in df_gold.columns:
    df_gold["Date"] = pd.to_datetime(df_gold["Date"])
    df_gold.set_index("Date", inplace=True)
else:
    if not pd.api.types.is_datetime64_any_dtype(df_gold.index):
        df_gold.index = pd.to_datetime(df_gold.index)

# Limpiar columnas numéricas de comas y convertir a float
for col in df_gold.columns:
    df_gold[col] = df_gold[col].astype(str).str.replace(',', '')
    df_gold[col] = pd.to_numeric(df_gold[col], errors='coerce')

print("Dimensiones del dataset:", df_gold.shape)
print("Columnas disponibles:\n", df_gold.columns)
print("Fechas disponibles:", df_gold.index.min(), "→", df_gold.index.max())
print("Tipos de datos por columna:\n", df_gold.dtypes)

print("Estadísticas descriptivas por divisa:")
display(df_gold.describe().T)

plt.figure(figsize=(14, 8))
sns.heatmap(df_gold.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación entre precios del oro en diferentes divisas")
plt.show()

plt.figure(figsize=(15, 6))
for col in df_gold.columns[:5]:
    plt.plot(df_gold[col], label=col)
plt.legend()
plt.title("Evolución del precio del oro en las primeras 5 divisas")
plt.ylabel("Precio")
plt.grid(True)
plt.show()

rolling_std = df_gold.rolling(window=30).std()
plt.figure(figsize=(15, 6))
for col in df_gold.columns[:5]:
    plt.plot(rolling_std[col], label=f"{col} (30d std)")
plt.legend()
plt.title("Volatilidad del oro (30 días de media móvil)")
plt.grid(True)
plt.show()

df_gold.pct_change().hist(bins=100, figsize=(16, 8), sharex=True)
plt.suptitle("Distribución de retornos diarios del oro por divisa")
plt.show()
```
